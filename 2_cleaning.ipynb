# DGL 

- **Framework Agnostic**:  
    Build your models with PyTorch, TensorFlow or Apache MXNet.
- **Efficient And Scalable**:  
    Fast and memory-efficient message passing primitives for training Graph Neural Networks. Scale to giant graphs via multi-GPU acceleration and distributed training infrastructure.

# How to create Graph in DGL
- DGL works with local ids, so you need to create an autoincrement id starting from zero for each node type.
- You need canonical edge types composed of (source node, edge type, target node) and to create relations for an edge you have to create a tensor of source nodes and a parallel tensor of target nodes.
- DGL works with tensor as node features so you need to convert string, categorical, ecc

Useful query to check ratio
```
MATCH (n:Business)--(c:Category)  
WITH c.name as name, collect(COALESCE(n.attributes_ByAppointmentOnly, "null")) as allapp  
RETURN name, 
    size([x in allapp WHERE x = "null"]) as null1,
    size([x in allapp WHERE x = "True"]) as True1,
    size([x in allapp WHERE x = "False"]) as False1,
    toFloat(size([x in allapp WHERE x = "null"])) / size(allapp) as ratio ORDER BY ratio DESC
```

# TIP

````python
import pandas as pd
import torch
tip_df = pd.read_csv('neo4j_csvs/tip.csv')
````

````python
tip_df
````

````python
tip_df.fillna('', inplace=True)
````

````python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("LiYuan/amazon-review-sentiment-analysis")
model = AutoModelForSequenceClassification.from_pretrained("LiYuan/amazon-review-sentiment-analysis")
````

````python
from datasets import Dataset
dataset = Dataset.from_pandas(tip_df[['text']])
dataset = dataset.map(lambda examples: model(**tokenizer(examples["text"], padding=True, return_tensors="pt")), batched=True, batch_size=1024)
````

````python
tip_logits_stars = torch.tensor(dataset['logits'], dtype=torch.float64)
torch.save(tip_logits_stars, 'preprocessed/tip_logits_stars.pt')
````

````python
tip_logits_stars = torch.load('preprocessed/tip_logits_stars.pt')
````

````python
tip_logits_stars.shape
````

````python
tip_compliment = torch.tensor(tip_df['compliment_count'].values, dtype=torch.float64)
````

````python
tip_compliment.sum()
````

````python
tip_features = torch.cat((tip_compliment[:, None], tip_logits_stars), axis=1)
````

````python
tip_features.shape
````

````python
torch.save(tip_features, 'preprocessed/tip_features.pt')
````

````python
tip_df['tip_id:ID'].to_csv('preprocessed/tip_ids.csv', index=False)
````
