```python
float_columns = ['latitude', 'longitude', 'stars', 'review_count', 'is_open', 'attributes_BusinessAcceptsCreditCards', 'attributes_RestaurantsPriceRange2']
list_of_dummies_columns = list(set(business_df.columns) - set(['business_id:ID', *float_columns]))
```

```python
dummies
```

```python
floats = business_df[float_columns].fillna(-1).astype(float)
```

```python
business_features = pd.concat([floats, dummies], axis=1).astype(float)
```

```python
business_features
```

```python
business_ids = business_df['business_id:ID']
```

```python
business_tensor_features = torch.tensor(business_features.values, dtype=torch.float64)
```

```python
torch.save(business_tensor_features, 'preprocessed/business_feature.pt')
```

```python
business_ids.to_csv('preprocessed/business_ids.csv', index=False)
```

# TIP

```python
import pandas as pd

tip_df = pd.read_csv('neo4j_csvs/tip.csv')
```

```python
tip_df
```

```python
tip_df.fillna('', inplace=True)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("LiYuan/amazon-review-sentiment-analysis")
model = AutoModelForSequenceClassification.from_pretrained("LiYuan/amazon-review-sentiment-analysis")
```

```python
inputs = tokenizer("Hello world!", return_tensors="pt")
outputs = model(**inputs)
```

```python
outputs.logits
```

```python
from datasets import Dataset
dataset = Dataset.from_pandas(tip_df[['text']])
dataset.map(lambda examples: model(**tokenizer(examples["text"], padding=True, return_tensors="pt")), batched=True, batch_size=256)
```
